# **merge schema**

```python
new_df = df.withColumn('discount',lit(0.1))\
            .withColumn('delivery_date', current_date())


new_df.write.format('delta')\
    .mode('append')\
    .option('mergeSchema',True)\
    .saveAsTable('parent_cata.delta_lake.retail_orders')
```

# **overwriteschema**

```python
df = spark.read.table('parent_cata.delta_lake.retail_orders')

df.write.format('delta')\
    .mode('overwrite')\
    .saveAsTable('parent_cata.delta_lake.retail_orders_copy')
```

```python
df = spark.read.table('parent_cata.delta_lake.retail_orders_copy')

df.display()
```

```python
new_schema = StructType([
    StructField("order_id", IntegerType(), True),
    StructField("customer_id", IntegerType(), True),
    StructField("product", StringType(), True),
    StructField("category", StringType(), True),
    StructField("quantity", IntegerType(), True),
    StructField("price", DoubleType(), True),
    StructField("status", StringType(), True),
    StructField("region", StringType(), True)
])


data_new = [
    (201, 501, "Keyboard", "Electronics", 2, 1800.0, "Delivered", "South"),
    (202, 502, "Mouse", "Electronics", 1, 900.0, "Shipped", "North"),
]
df_overwrite = spark.createDataFrame(data_new, schema=new_schema)

# Overwrite and replace schema
df_overwrite.write.format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("parent_cata.delta_lake.retail_orders_copy")

df_overwrite.display(df)
```
