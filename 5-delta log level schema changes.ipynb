```python
data = [
    ("Alice", 25, "F", "NY"),
    ("Bob", 30, "M", "CA"),
    ("Carol", 28, "F", "TX"),
    ("Dave", 35, "M", "WA")
]
columns = ["Name", "Age", "Gender", "State"]
df = spark.createDataFrame(data, columns)
display(df)
```
