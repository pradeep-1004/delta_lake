# **upsert**

```python
from delta.tables import DeltaTable

df_table2_delta = DeltaTable.forName(spark, 'parent_cata.delta_lake.dml_2')




df_table2_delta.alias('target').merge(
    df_table1.alias('source'),
    'target.customer_id = source.customer_id'
).whenMatchedUpdateAll(
).whenNotMatchedInsertAll(
).execute()
```

```python
spark.table("parent_cata.delta_lake.dml_2").orderBy("customer_id").display()
```
